```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

token = word_tokenize(parag)

stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in token if word.lower() not in stop_words]

lemmatizer = WordNetLemmatizer()
lema_word = [lemmatizer.lemmatize(word) for word in filtered_tokens]

lema_sen = " ".join(lema_word)
print(lema_sen)
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import string

lemmatizer = WordNetLemmatizer()
token = word_tokenize(parag)

stop_words = set(stopwords.words('english'))

filtered_tokens = [
    w for w in token
    if w.lower() not in stop_words and w not in string.punctuation
]

lema_word = [lemmatizer.lemmatize(w, pos='v') for w in filtered_tokens]
lema_sen = " ".join(lema_word)

lema_sen
```
